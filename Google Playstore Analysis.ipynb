## Data Cleaning

```python
for i in df.columns:
    col = i.replace(' ','')
    df = df.rename(columns={i:col})
df.columns   
```

```python
df.drop(['AppId','Currency','DeveloperWebsite','PrivacyPolicy','ScrapedTime'], axis =1, inplace = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
NullCol = df.columns[df.isnull().any()]

for i in NullCol:
    print(i , ':' , df[i].isnull().sum())
    
    
```

```python
df.dropna(subset=['AppName','Installs' ,'MinimumInstalls','Size'], inplace =True)
```

```python
df.Installs = df.Installs.str.replace(',','')
df.Installs = df.Installs.str.replace('+','')
df.Installs = pd.to_numeric(df.Installs)
```

```python
df['Rating'] = df['Rating'].fillna(value= df['Rating'].mean())
df['RatingCount'] = df['RatingCount'].fillna(value= df['RatingCount'].mean())
```

```python
def categorize_rating(i):
    if i >= 1 and i<= 1.9:
        return('1 Star')
    elif i >=2 and i<= 2.9:
        return('2 Star')
    elif i >=3 and i<= 3.9:
        return('3 Star')
    elif i >=4 and i<= 4.9:
        return('4 Star')
    elif i ==5:
        return('5 Star')
    else:
        return('No Ratings')
```

```python
df['Rating(Group)'] = df['Rating'].apply(categorize_rating)
```

```python
df
```

```python
df['Rating(Group)'].value_counts()
```

```python
df['ContentRating'].value_counts()
```

```python
df['ContentRating'] = df['ContentRating'].replace('Mature 17+',"Adults").replace('Adults only 18+','Adults').replace('Everyone 10+', 'Everyone').replace('Unrated', 'Everyone')
```

```python
df['ContentRating'].value_counts()
```

```python
df['PriceRange']= pd.cut(df['Price'],bins=[0,0.19,9.99,29.99,400],labels=['Free','Low','Mid','High'],include_lowest=True)
```

```python
df['Price'].value_counts()
```

```python
df
```

```python
df.loc[(df['Price']==0) & (df['Free']==False), 'Free']= True
df['Free'].value_counts()
```

```python
df['Free'] = np.where(df['Free']==True, 'Free', 'Paid')
df
```

```python
Numeric_Feature = [i for i in df.columns if df[i].dtype != 'O']
Category_Feature = [i for i in df.columns if df[i].dtype == 'O']

print('We have {} Numeric Features: {}' .format(len(Numeric_Feature),Numeric_Feature))
print('We have {} Category Features: {}' .format(len(Category_Feature),Category_Feature))      
```
